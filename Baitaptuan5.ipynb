Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> int:

```python
def max(a,b):
    return a if a > b else b
a = 2
b = 5
print(max(a,b))
```

W5A2 Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]: 


```python
def swap(a: int, b : int) -> tuple[int,int]:
    a,b = b,a
    return a, b
a = 4
b = 6
a, b = swap(a,b)
print(a,b)
```

W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
def so_nt(n:int) -> bool:
    if n < 2:
        return False
    elif n == 2:
        return True
    elif n % 2 == 0:
        return False
    for i in range(3,int(n**0.5)+1,2):
        if n % i == 0:
            return False
```
