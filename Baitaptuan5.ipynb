W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
def is_prime(n: int) -> bool:
    if n < 2: return False
    for i in range(2, int(n ** 0.5 + 1)):
        if n % i == 0:
            return False
    return True
n = int(input())
print(is_prime(n))
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def is_perfect_number(n: int) -> bool:
    if n < 2: return False
    sum_divisors = 1
    for i in range(2, int(n ** 0.5+ 1)):
        if n % i == 0:
            sum_divisors += i
            if i*i != n:
                sum_divisors += n // i
    return sum_divisors == n
n = int(input())
print(is_perfect_number(n))
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

```python
def find_first_index(lst: list, k: int) -> int:
    for i in range(len(lst)):
        if lst[i] == k:
            return i  
            
    return -1
lst = list(map(int, input().split()))
k = int(input())
print(find_first_index(lst, k))
```


W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def factorial(n: int) -> int:
    if n < 0:
        return 0 
    result = 1
    for i in range(1, n + 1):
        result = result * i
        
    return result
n = int(input())
print(factorial(n))
```

Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
\)
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
Phép trừ \( - \)
Phép cộng \( + \)
Phép chia \( / \)
Phép nhân \( * \)
Kết quả làm tròn đến chữ số thập phân thứ hai.
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
Đầu ra: Kết quả phép toán. 

```python
def simple_calculator(num1: float, num2: float, operator: str):
    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/' or operator == '^': 
        if num2 != 0:
            result = num1 / num2
        else:
            return "Lỗi: Không thể chia cho 0"
    else:
        return "Toán tử không hợp lệ"
    
    return round(result, 2)
print(simple_calculator(10,3,'/'))
```

W5A8
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 


```python
def hamming_distance(x: int, y: int) -> int:
    different_bits = x ^ y
    distance = 0
    while different_bits > 0:
        if different_bits & 1:
            distance += 1
        
        different_bits = different_bits >> 1
            
    return distance

x,y = map(int,input().split())
print(hamming_distance(x,y))
```

W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 


```python
def sum_of_digits(n: int) -> int:
    total = 0
    while n > 0:
        total += n % 10
        n = n // 10
    return total
n = int(input())
print(sum_of_digits(n))
```

W5A16
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số


```python
def is_armstrong(x: int) -> bool:
    s = str(abs(x))
    n = len(s)
    total = sum(int(digit) ** n for digit in s)
    return total == x

val = int(input())
print(is_armstrong(val))
```

```python
import math

def is_perfect_square(x: int) -> bool:
    if x < 0:
        return False
    root = math.isqrt(x)
    return root * root == x

try:
    num = int(input("Nhập vào một số nguyên x: "))
    
    if is_perfect_square(num):
        print(f"Kết quả: {num} LÀ số chính phương.")
    else:
        print(f"Kết quả: {num} KHÔNG PHẢI là số chính phương.")
        
except ValueError:
    print("Lỗi: Vui lòng nhập một số nguyên hợp lệ!")
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_palindrome(n: int) -> bool:
    s = str(abs(n))
    return s == s[::-1]

val = int(input())
print(is_palindrome(val))
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_magic_number(n: int) -> bool:
    n = abs(n)
    while n >= 10:
        total = 0
        for digit in str(n):
            total += int(digit)
        n = total
    
    return n == 1

try:
    num = int(input("Nhập vào số nguyên x: "))
    
    if is_magic_number(num):
        print(f"Số {num} LÀ số kỳ diệu.")
    else:
        print(f"Số {num} KHÔNG PHẢI là số kỳ diệu.")
        
except ValueError:
    print("Vui lòng nhập một số nguyên!")
```
