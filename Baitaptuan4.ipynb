VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi.
Kiểm tra số này có phải là số nguyên tố hay không 

```python
while True:
    try:
        n = int(input("Nhập một số nguyên dương: "))
        if n > 0:
            break
        else:
            print("Số phải lớn hơn 0!")
    except:
        print("Bạn phải nhập số nguyên!")

# Kiểm tra số nguyên tố
if n < 2:
    print(n, "không phải là số nguyên tố")
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print(n, "không phải là số nguyên tố")
            break
    else:
        print(n, "là số nguyên tố")
```

với 0 < n < 100.

W4A6

Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
a = int(input())
b = int(input())
tong = 0

for num in range(a, b + 1):
    # Kiểm tra num có phải nguyên tố không
    if num < 2:
        continue
    is_p = True
    j = 2
    while j * j <= num:
        if num % j == 0:
            is_p = False
            break
        j += 1
    
    if is_p:
        tong += num
print(tong)
```

Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 

```python
n = int(input("Nhập n (n >= 2): "))
max_prime = -1
d = 2

while d * d <= n:
    while n % d == 0:
        max_prime = d
        n //= d
    d += 1
if n > 1:
    max_prime = n 

print(max_prime)
```

W4A8
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu 

```python
n = int(input("Nhập n: "))
steps = 0

while True:
    temp = n
    rev_n = 0
    while temp > 0:
        rev_n = rev_n * 10 + temp % 10
        temp //= 10
    
    if n == rev_n:
        break # Dừng vòng lặp
    n = n + rev_n
    steps += 1

print(f"{steps} {n}")
```

[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number)

```python
# WW4A9
n = int(input("Nhập n: "))
found = False

i = 1
while i * i <= n:
    sq = i * i
    
    # Kiểm tra các chữ số của sq có khác nhau không
    temp = sq
    is_distinct = True
    # Dùng 2 vòng lặp lồng nhau để so sánh các chữ số (vì ko được dùng mảng/chuỗi)
    # Cách đơn giản hơn: Dùng bitmask (nâng cao) hoặc tách số
    # Ở đây mình dùng cách tách số thủ công:
    
    mask = 0 # Dùng 1 số nguyên làm bộ nhớ (bitmask)
    temp_check = sq
    while temp_check > 0:
        digit = temp_check % 10
        if (mask >> digit) & 1: # Nếu bit thứ digit đã bật -> trùng
            is_distinct = False
            break
        mask = mask | (1 << digit) # Bật bit thứ digit lên
        temp_check //= 10
        
    if is_distinct:
        print(sq, end=" ")
        found = True
    i += 1

if not found:
    print("no number")
else:
    print() # Xuống dòng
```

W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python
# W4A10
n = int(input("Nhập n: "))
max_len = 0
best_num = 1

for x in range(1, n + 1):
    curr = x
    length = 1
    while curr != 1:
        if curr % 2 == 0:
            curr //= 2
        else:
            curr = 3 * curr + 1
        length += 1
    
    if length > max_len:
        max_len = length
        best_num = x
    # Nếu bằng nhau thì giữ x nhỏ hơn (do x chạy từ 1 lên nên tự thỏa mãn)

print(f"{best_num} {max_len}")
```
